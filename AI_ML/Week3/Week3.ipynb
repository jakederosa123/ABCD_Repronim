```python
n_points = 1000
X, color = datasets.make_s_curve(n_points, random_state=0)
n_neighbors = 10
n_components = 2

# Create figure
fig = plt.figure(figsize=(45, 24))

# Add 3d scatter plot
ax = fig.add_subplot(251, projection="3d")
ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral)
ax.view_init(4, -72)
```

```python
'''
What sort of scenarios are they typically used in?

#################################################
Manifold Learning methods
#################################################

# 1) Isomap 
This is a good choice to use for dimensionality reduction while preserving non-linear relatioships.
Often used in digit recognition and natural language processing.

# 2) Spectral 
sklearn notes: very useful when the structure of the individual clusters is highly non-convex, or more 
generally when a measure of the center and spread of the cluster is not a suitable description 
of the complete cluster, such as when clusters are nested circles on the 2D plane.

# 3) Multidimensional scaling
This a good choice for understanding the dissimilarity of multivariate patterns in a highdimensional space. 
For example if I am looking to investigate the multivariate patterns across different working operations within 
functional network, MDS may be a good choice. 
'''
```

```python
# Isomap embedding
iso = manifold.Isomap(n_neighbors=n_neighbors,n_components=n_components)
iso_map = iso.fit_transform(X)
fig = plt.figure(figsize=(40, 15))
ax = fig.add_subplot(2, 5, 2)
ax.scatter(iso_map[:, 0], iso_map[:, 1], c=color, cmap=plt.cm.Spectral)
ax.axis("tight")
```

```python
# Spectral Embedding
spectral = manifold.SpectralEmbedding(n_neighbors=n_neighbors,n_components=n_components)
spectral_map = spectral.fit_transform(X)
fig = plt.figure(figsize=(40, 15))
ax = fig.add_subplot(2, 5, 2)
ax.scatter(spectral_map[:, 0], spectral_map[:, 1], c=color, cmap=plt.cm.Spectral)
ax.axis("tight")
```

```python
# Local Linear Embedding - LTSA
local = manifold.LocallyLinearEmbedding(method="ltsa",n_neighbors=n_neighbors,n_components=n_components, eigen_solver="auto")
local_map = local.fit_transform(X)
fig = plt.figure(figsize=(40, 15))
ax = fig.add_subplot(2, 5, 2)
ax.scatter(local_map[:, 0], local_map[:, 1], c=color, cmap=plt.cm.Spectral)
ax.axis("tight")
```

```python
# MDS
mds = manifold.MDS(n_components=n_components, max_iter=100, n_init=1)
mds_map = mds.fit_transform(X)
fig = plt.figure(figsize=(40, 15))
ax = fig.add_subplot(2, 5, 2)
ax.scatter(mds_map[:, 0], mds_map[:, 1], c=color, cmap=plt.cm.Spectral)
ax.axis("tight")
```

```python
# Create an extractor which will allow us to create ICA network timeseries
extractor = RegionExtractor(canica_components_img, 
                            threshold=0.5,
                            thresholding_strategy='ratio_n_voxels',
                            extractor='local_regions',
                            standardize=True, 
                            min_region_size=1350)

print(f'Fitting the Extactor...')

# Fit the extractor
extractor.fit()

# Extracted regions are stored in regions_img_
regions_extracted_img = extractor.regions_img_

# Total number of regions extracted
n_regions_extracted = regions_extracted_img.shape[-1]

print(f'{n_regions_extracted} Regions extracted...')
```

```python
# Store sub (key), and adjacency matrix (val)
mat_dict = {}

# ConnectivityMeasure
connectome_measure = ConnectivityMeasure(kind='correlation')
func_filenames, confounds = rest_dataset.func, rest_dataset.confounds

# Loop through each subject
for filename, confound in zip(func_filenames, confounds):

    # Get cleaner version of subject name
    sub = filename.split(os.path.sep)[-1]
    sub = sub.split('_')[0]
    
    print(f'Creating correlation matrix for: {sub}')

    # Extract timeseries signals from our extractor
    timeseries_each_subject = extractor.transform(filename, confounds=confound)
    
    # Create our Connectivity Measure from this extracted time series
    mat = connectome_measure.fit_transform([timeseries_each_subject])
    mat = mat.squeeze(axis = 0)
    
    # Add to our dict
    mat_dict[sub] = mat
```

```python
plt.imshow(mat, cmap = 'coolwarm')
```

```python
num_subs = len(mat_dict.keys())
x, y = mat.shape
num_feats = int(((x * y) - x) / 2)

mat_full = np.zeros((num_subs, num_feats))

# Generate an emptry matrix to store this data
for ind, (sub, mat) in enumerate(mat_dict.items()):
    mat_full[ind, :] = mat[np.triu_indices(mat.shape[0], k = 1)]
    
print(mat_full.shape)
       
fig, ax = plt.subplots(figsize = (8, 4))
ax.imshow(mat_full, aspect = 'auto', cmap = 'coolwarm')
plt.show()
```

```python
X = mat_full

pheno_data = rest_dataset['phenotypic']
Y = np.array([row[1] for row in pheno_data])

print(f'X shape: {X.shape} Y shape: {Y.shape}')
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split, KFold, cross_validate
from sklearn.metrics import r2_score
```

```python
# Create the test train split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)

X_train
```

```python
# Create and fit the model
model = Ridge()
model.fit(X_train, y_train)

# Predict on our training data and plot (actual vs pred) and print the score, mse
pred = model.predict(X_train)
plt.scatter(y_train, pred.squeeze())

print(f'Model Score: {r2_score(y_train, pred)}')
```

```python
# Predict on our training data and plot (actual vs pred) and print the score, mse
pred = model.predict(X_test)
plt.scatter(y_test, pred.squeeze())

print(f'Model Score: {r2_score(y_test, pred)}')
```

```python
import networkx as nx
```

```python
# Store our graphs for each subject
graphs_dict = {}

for key, val in mat_dict.items():
    
    # Apply thresholding and binarization
    val = np.abs(val)
    val[val > .2] = 1
    val[val < 1] = 0
    
    # Store graph
    g = nx.from_numpy_matrix(val)
    graphs_dict[key] = g
    
```

```python
nx.draw_kamada_kawai(g)
```

```python
#Average clustering coefficient
acg_clust = nx.average_clustering(g)
print(f'Average clustering coefficient: {round(acg_clust,2)}')
```

```python
#Global efficiency
glb_eff = nx.global_efficiency(g)
print(f'Global efficiency: {round(glb_eff,2)}')
```

```python
#Average degree
avg_deg = np.mean(nx.degree(g))
print(f'Global efficiency: {round(avg_deg,2)}')
```

```python
gMetric_dict = {'clust_co': [],
                'efficiecy': [],
                'degree': []}

for sub, g in graphs_dict.items():
    gMetric_dict['clust_co'].append(nx.average_clustering(g))
    gMetric_dict['efficiecy'].append(nx.global_efficiency(g))
    gMetric_dict['degree'].append(np.mean(nx.degree(g)))
```

```python
fig, ax = plt.subplots(1, 3, figsize = (12, 4))

ax[0].scatter(gMetric_dict['clust_co'], gMetric_dict['efficiecy'])
ax[1].scatter(gMetric_dict['clust_co'], gMetric_dict['degree'])
ax[2].scatter(gMetric_dict['efficiecy'], gMetric_dict['degree'])
```
